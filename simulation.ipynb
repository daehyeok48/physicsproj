```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
```

# freq_I

```python
# 회로 매개변수
L = 1.0  # 인덕턴스 (H)
C = 1.0  # 커패시턴스 (F)
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
E_ext = 1.0  # 외부 전압 진폭 (V)

# 주파수 범위 설정
omega = np.linspace(0.1, 10, 500)  # 주파수 (rad/s)

# 주파수 응답 계산
plt.figure(figsize=(12, 8))

for R in R_values:
    # 임피던스 계산
    Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C)) ** 2)  # 임피던스 크기
    I = E_ext / Z  # 전류 크기
    plt.plot(omega, I, label=f"R = {R} Ω", linewidth=2)

# 그래프 꾸미기
plt.title("Frequency Response of RLC Circuit", fontsize=24)
plt.yscale("log")
plt.xlabel("Frequency (ω)", fontsize=20)
plt.ylabel("Amplitude of Current (I)", fontsize=20)
plt.legend(fontsize=18)
plt.xlim(0, 10)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(True)
plt.show()
```

# 위상자 차이 _ C 0.1,1

```python
# 파라미터 정의
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
L = 1  # 인덕턴스 (헨리)
C_values = [1, 0.1]
lines = ["--", "-"]
E_ext = 1  # 외부 기전력 (볼트)
omega = 1  # 각주파수 (라디안/초)
colors = ["#FF6347", "#800080", "#4169E1", "#0000FF", "#3CB371", "#006400"]

# 시간 설정
t = np.linspace(0, 50, 1000)  # 0초에서 50초까지 1000개의 데이터 포인트

plt.figure(figsize=(24, 12))

# C=1 I(t)를 점선 R에 따라
# C=0.1 I(t)를 실선 R에 따라

for i in [0, 1]:
    for j in [0, 1, 2, 3, 4]:

        # temp

        C_temp = C_values[i]
        R_temp = R_values[j]
        # 공진 주파수 계산
        omega_0 = 1 / np.sqrt(L * C_temp)

        # 임피던스 계산
        Z = np.sqrt(R_temp**2 + (omega * L - 1 / (omega * C_temp)) ** 2)

        # 전류 진폭 계산
        I_0 = E_ext / Z

        # 시간에 따른 전류 계산
        phi = np.arctan((omega * L - 1 / (omega * C_temp)) / R_temp)  # 위상각
        # print(phi)
        I_t = I_0 * np.sin(omega * t - phi)  # I(t)

        # x축을 w*t로 설정
        wt = omega * t

        # 그래프 그리기
        plt.plot(
            wt,
            I_t,
            label=f"R={R_temp}, C={C_temp}",
            color=colors[j],
            linestyle=lines[i],
        )

        print(f"phi={phi:.2f} at R={R_temp}, C={C_temp}")

plt.title("Current in RLC Circuit Over Time", fontsize=30)
plt.xlabel("$\omega t$", fontsize=26)
plt.ylabel("Amplitude of Current (I)", fontsize=26)
plt.axhline(0, color="black", linewidth=2.0, linestyle="-")
plt.legend(fontsize=20, loc="lower right")
plt.xlim(0, 8)
plt.ylim(-0.25,0.25)
plt.xticks(fontsize=26)
plt.yticks(fontsize=24)
plt.grid(True)
plt.show()
```

```python
# 파라미터 정의
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
colors = ["#FF6347", "#800080", "#4169E1", "#0000FF", "#3CB371", "#006400"]
L = 1  # 인덕턴스 (헨리)
C = 1  # 커패시턴스 (패럿)
omega = 1  # 각주파수 (라디안/초)

plt.figure(figsize=(8, 8))

for i in [4, 3, 2, 1, 0]:
    R = R_values[i]
    # 벡터 구성
    R_vector = np.array([R, 0])  # 저항 벡터
    L_vector = np.array([0, omega * L])  # 인덕턴스 벡터 (jωL)
    C_vector = np.array([0, -1 / (omega * C)])  # 커패시턴스 벡터 (-j/(ωC))

    # 총 임피던스 벡터 계산
    Z_vector = R_vector + L_vector + C_vector

    # 그래프 그리기

    plt.quiver(
        0,
        0,
        R_vector[0],
        R_vector[1],
        angles="xy",
        scale_units="xy",
        scale=1,
        color=colors[i],
        label=f"R = {R}",
    )
    plt.quiver(
        0,
        0,
        L_vector[0],
        L_vector[1],
        angles="xy",
        scale_units="xy",
        scale=1,
        color=colors[i],
    )
    plt.quiver(
        0,
        0,
        C_vector[0],
        C_vector[1],
        angles="xy",
        scale_units="xy",
        scale=1,
        color=colors[i],
    )
    plt.quiver(
        0,
        0,
        Z_vector[0],
        Z_vector[1],
        angles="xy",
        scale_units="xy",
        scale=2,
        color=colors[i],
        label=f"Z of R = {R}",
    )


# 축 설정
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
# plt.xscale("log")
# plt.yscale("log")
plt.xlabel("Real Part (Ω)", fontsize=16)
plt.ylabel("Imaginary Part (Ω)", fontsize=16)
plt.title("Phasor Diagram of RLC Circuit at L,C=1", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

# inductor energy

```python
# 파라미터 정의
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
L = 1  # 인덕턴스 (헨리)
C = 1
lines = ["--", "-"]
E_ext = 1  # 외부 기전력 (볼트)
omega = 1  # 각주파수 (라디안/초)
colors = ["#FF6347", "#800080", "#4169E1", "#0000FF", "#3CB371", "#006400"]

# 시간 설정
t = np.linspace(0, 50, 1000)  # 0초에서 50초까지 1000개의 데이터 포인트

plt.figure(figsize=(24, 12))

for R in R_values:

    # 공진 주파수 계산
    omega_0 = 1 / np.sqrt(L * C)

    # 임피던스 계산
    Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C)) ** 2)

    # 전류 진폭 계산
    I_0 = E_ext / Z

    # 시간에 따른 전류 계산
    phi = np.arctan((omega * L - 1 / (omega * C)) / R)  # 위상각
    # print(phi)
    I_t = I_0 * np.sin(omega * t - phi)  # I(t)

    # 인덕터 자기에너지 계산
    Inductor = (1 / 2) * L * (I_t**2)

    # x축을 w*t로 설정
    wt = omega * t

    # 그래프 그리기
    plt.plot(
        wt,
        Inductor,
        label=f"R={R}, C={C}",
        linestyle=lines[i]
    )

plt.title("Energy of inductor in RLC Circuit Over Time", fontsize=30)
plt.xlabel("$\omega t$", fontsize=26)
plt.ylabel("Energy of Inductor", fontsize=26)
plt.axhline(0, color="black", linewidth=2.0, linestyle="-")
plt.legend(fontsize=20, loc="lower right")
plt.xlim(0, 8)
plt.yscale("log")
plt.xticks(fontsize=26)
plt.yticks(fontsize=24)
plt.grid(True)
plt.show()
```

```python
# 파라미터 정의
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
L = 1  # 인덕턴스 (헨리)
C = 1
lines = ["--", "-"]
E_ext = 1  # 외부 기전력 (볼트)
omega = 1  # 각주파수 (라디안/초)
colors = ["#FF6347", "#800080", "#4169E1", "#0000FF", "#3CB371", "#006400"]

# 시간 설정
t = np.linspace(0, 50, 1000)  # 0초에서 50초까지 1000개의 데이터 포인트

V=E_ext*np.cos(omega*t)

plt.figure(figsize=(24, 12))

for R in R_values:

    # 공진 주파수 계산
    omega_0 = 1 / np.sqrt(L * C)

    # 임피던스 계산
    Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C)) ** 2)

    # 전류 진폭 계산
    I_0 = E_ext / Z

    # 시간에 따른 전류 계산
    phi = np.arctan((omega * L - 1 / (omega * C)) / R)  # 위상각
    # print(phi)
    I_t = I_0 * np.sin(omega * t - phi)  # I(t)

    # V
    V=E_ext*np.cos(omega*t)+phi

    # 전력 계산
    Power = (1/2)*I_t*V*np.cos(phi)

    # x축을 w*t로 설정
    wt = omega * t

    # 그래프 그리기
    plt.plot(
        wt,
        Power,
        label=f"R={R}, C={C}"
    )

plt.title("Power in RLC Circuit Over Time", fontsize=30)
plt.xlabel("$\omega t$", fontsize=26)
plt.ylabel("Power", fontsize=26)
plt.axhline(0, color="black", linewidth=2.0, linestyle="-")
plt.legend(fontsize=20, loc="lower right")
plt.xlim(-0.5, 10)
#plt.yscale("log")
plt.xticks(fontsize=26)
plt.yticks(fontsize=24)
plt.grid(True)
plt.show()
```

```python


# 비선형 유리 함수 정의 (분모에 R^2 포함)
def rational_function(x, a, b, c):
    return a / (b * x**2 + c)

# 파라미터 정의
R_values = np.array([0.01, 0.1, 1.0, 10.0, 100.0])  # 저항값 (Ω)
VI_values = []  # 전력 저장

# 시뮬레이션 데이터 생성 (예제 데이터)
E_ext = 1  # 외부 전압 (V)
omega = 1  # 각주파수 (rad/s)
L = 1      # 인덕턴스 (H)
C = 1    # 커패시턴스 (F)
t = np.linspace(0, 1, 1000)  # 시간 벡터

for R in R_values:
    V_data=[]
    Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C)) ** 2)
    I_0 = E_ext / Z
    phi = np.arctan((omega * L - 1 / (omega * C)) / R)
    I_t = I_0 * np.sin(omega * t - phi)
    I_rms = np.sqrt(np.mean(np.square(I_t)))
    V=E_ext*np.cos(omega*t)+phi
    V_data.append(V)
    V_rms = np.sqrt(np.mean(np.square(V_data)))
    VI = I_rms * V_rms
    VI_values.append(VI)

# 데이터를 배열로 변환
VI_values = np.array(VI_values)

# 비선형 유리 함수 피팅
params, _ = curve_fit(rational_function, R_values, VI_values)

# 피팅된 함수의 식 출력
a, b, c = params
print(f"피팅된 유리 함수는 다음과 같습니다:")
print(f"f(x) = ({a:.2f}) / ({b:.2f}x^2 + {c:.2f})")

# 피팅된 함수로 데이터 예측
R_fit = np.linspace(min(R_values), max(R_values), 100)
VI_fit = rational_function(R_fit, *params)

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.scatter(R_values, VI_values, color='blue', label='Measured Data', s=150)
plt.plot(R_fit, VI_fit, color='red', linewidth=2, label='Rational Function Fit')

plt.title("RMS Power in RLC Circuit Over Resistance with Rational Function Fit", fontsize=20)
plt.xlabel("Resistance (Ω)", fontsize=16)
plt.ylabel("V_rms * I_rms (W)", fontsize=16)
plt.legend(fontsize=14, loc="upper left")
plt.grid(True)
plt.show()
```

```python
# 파라미터 정의
R_values = [0.01, 0.1, 1.0, 10.0, 100.0]  # 저항값 (Ω)
L = 1  # 인덕턴스 (헨리)
C = 1
lines = ["--", "-"]
E_ext = 1  # 외부 기전력 (볼트)
omega = 1  # 각주파수 (라디안/초)
colors = ["#FF6347", "#800080", "#4169E1", "#0000FF", "#3CB371", "#006400"]

# 시간 설정
t = np.linspace(0, 50, 1000)  # 0초에서 50초까지 1000개의 데이터 포인트

plt.figure(figsize=(24, 12))

for R in R_values:

    # 공진 주파수 계산
    omega_0 = 1 / np.sqrt(L * C)

    # 임피던스 계산
    Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C)) ** 2)

    # 전류 진폭 계산
    I_0 = E_ext / Z

    # 시간에 따른 전류 계산
    phi = np.arctan((omega * L - 1 / (omega * C)) / R)  # 위상각
    # print(phi)
    I_t = I_0 * np.sin(omega * t - phi)  # I(t)
    
    # V
    V=E_ext*np.cos(omega*t)+phi
    
    # 정전에너지
    E = -1*(V*I_t/(4*omega))*np.cos(2*omega*t)

    # x축을 w*t로 설정
    wt = omega * t

    # 그래프 그리기
    plt.plot(
        wt,
        E,
        label=f"R={R}, C={C}"
    )

plt.title("Capacity Energy in RLC Circuit Over Time", fontsize=30)
plt.xlabel("$\omega t$", fontsize=26)
plt.ylabel("Capacity Energy", fontsize=26)
plt.axhline(0, color="black", linewidth=2.0, linestyle="-")
plt.legend(fontsize=20, loc="lower right")
plt.xlim(-0.5, 10)
#plt.yscale("log")
plt.xticks(fontsize=26)
plt.yticks(fontsize=24)
plt.grid(True)
plt.show()
```
